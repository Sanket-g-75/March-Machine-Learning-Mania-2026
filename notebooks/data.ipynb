# Feature Engineering

```python
df = pd.read_csv(os.path.join(path,'data','processed','FData.csv'))
```

```python
df.columns
```

```python
df.isnull().sum() # No Null Values
```

```python
df['FGA3'] = df['FGA3'] + 0.001
df['FTA'] = df['FTA'] + 0.001
```

```python

df['FG%'] = df['FGM']/df['FGA']
df['FG%'].apply(lambda x: math.ceil(x * 10000) / 10000)

df['3P%'] = df['FGM3']/df['FGA3']
df['3P%'].apply(lambda x: math.ceil(x * 10000) / 10000)

df['FT%'] = df['FTM']/df['FTA']
df['FT%'].apply(lambda x: math.ceil(x * 10000) / 10000)
```

```python
def possesion(row):
    calc = (row['FGA'] - row['OR'] + row['TO']) + (0.44*row['FTA'])
    return calc
```

```python
df['Possessions'] = df.apply(lambda x: possesion(x),axis=1)
```

```python
df.columns
```

## Aggregation

```python
df = df.sort_values(['Season', 'DayNum', 'TeamID'])
```

```python
df['Win'] = (df['Score'] > df['OppScore']).astype(int)
```

```python
df.to_csv(os.path.join(path,'data','final','Matchups.csv'),index=False)
```

```python
grouped = df.groupby(['TeamID'])
```

```python
agg_df = grouped.agg({
    'Score': 'mean',
    'FGM': 'mean',
    'FGA': 'mean',
    'FGM3': 'mean',
    'FGA3': 'mean',
    'FTM': 'mean',
    'FTA': 'mean',
    'OR': 'mean',
    'DR': 'mean',
    'Ast': 'mean',
    'TO': 'mean',
    'Stl': 'mean',
    'Blk': 'mean',
    'PF': 'mean',
    'FG%': 'mean',
    '3P%': 'mean',
    'FT%': 'mean',
    'Possessions': 'mean',
    'NumOT': 'mean'
}).reset_index()
```

```python
agg_df['GamesPlayed'] = grouped.size().values
```

```python
win_rate = grouped['Win'].mean().reset_index(name='WinRate')
```

```python
agg_df = agg_df.merge(win_rate, on=['TeamID'])
```

```python
df['RollingScore'] = df.groupby(['TeamID'])['Score']\
    .transform(lambda x: x.rolling(10, min_periods=1).mean())
```

```python
recent = df.groupby(['TeamID'])['RollingScore'].last().reset_index()
```

```python
home = df[df['Loc'] == 'H'].groupby(['TeamID'])['Score'].mean()
away = df[df['Loc'] == 'A'].groupby(['TeamID'])['Score'].mean()
```

```python
final_team_df = agg_df.merge(recent, on=['TeamID'], how='left')
```

```python
final_team_df
```

```python
final_team_df = teams.merge(final_team_df, on="TeamID", how="left")
```

```python
final_team_df = final_team_df.fillna(0)
```

```python
final_team_df.to_csv(os.path.join(path,'data','final','Dataset.csv'),index=False)
```
