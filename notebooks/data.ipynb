```python
def ground(row):
    if row['WLoc'] == 'H':
        return 'A'
    elif row['WLoc'] == 'A':
        return 'H'
    else:
        return 'N'
    

data['LLoc'] = data.apply(ground,axis=1)
data2['LLoc'] = data2.apply(ground,axis=1)
```

```python
# first I will have to separate winners and losers
dataw = data[['Season', 'DayNum', 'WTeamID', 'WScore', 'WLoc',
       'NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF','Gender']]

datal = data[['Season', 'DayNum','LTeamID', 'LScore','LLoc',
              'NumOT', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR',
              'LAst', 'LTO', 'LStl', 'LBlk', 'LPF','Gender']]

dataw.rename(columns={'WTeamID':'TeamID'},inplace=True)
datal.rename(columns={'LTeamID':'TeamID'},inplace=True)
```
