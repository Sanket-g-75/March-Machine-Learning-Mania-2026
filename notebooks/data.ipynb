## Merging Men and Women Files

```python
# Teams

teams = pd.concat(
    [
    pd.read_csv(os.path.join(path,'data','raw','MTeams.csv')).assign(Gender='M'),
    pd.read_csv(os.path.join(path,'data','raw','WTeams.csv')).assign(Gender='W')
    ]
)
```

```python

teamconferences = pd.concat(
    [
    pd.read_csv(os.path.join(path,'data','raw','MTeamConferences.csv')).assign(Gender='M'),
    pd.read_csv(os.path.join(path,'data','raw','WTeamConferences.csv')).assign(Gender='W')
    ]
)
```

```python

ncaatourneyseeds = pd.concat(
    [
    pd.read_csv(os.path.join(path,'data','raw','MNCAATourneySeeds.csv')).assign(Gender='M'),
    pd.read_csv(os.path.join(path,'data','raw','WNCAATourneySeeds.csv')).assign(Gender='W')
    ]
)
```

```python

secondarytourneyteams = pd.concat(
    [
    pd.read_csv(os.path.join(path,'data','raw','MSecondaryTourneyTeams.csv')).assign(Gender='M'),
    pd.read_csv(os.path.join(path,'data','raw','WSecondaryTourneyTeams.csv')).assign(Gender='W')
    ]
)
```
